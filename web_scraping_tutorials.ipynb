```python
def myHackerNews(links, votes):
    hn = [] #empty list
    for idx, item in enumerate(links):
        title = links[idx].getText()
        href = links[idx].get('href')
        hn.append({"title" : title, "href": href})
    return hn
    
print(myHackerNews(links, votes))
```

```python
def myHackerNews(links, votes):
    hn = [] #empty list
    for idx, item in enumerate(links):
        title = links[idx].getText()
        
        href = links[idx].find_all('a', href=True)
        # Print out the result
        for link in href:
            if link.get('href') != None:
                if 'https://' in link.get('href'):
                    #print(link.get('href'))
                    href = link.get('href')
                    hn.append({"title" : title, "href": href})
        
    return hn
    
print(myHackerNews(links, votes))
```
